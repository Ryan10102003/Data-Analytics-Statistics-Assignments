#Assignment Questions


1. What is hypothesis testing in statistics?
 - Hypothesis testing is a method to make decisions using data, by testing an assumption (hypothesis) about a population parameter.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - The null hypothesis (H₀) assumes no effect or difference, while the alternative hypothesis (H₁) suggests there is an effect or difference.

3.  What is the significance level in hypothesis testing, and why is it important?
 - The significance level (α) is the probability of rejecting the null hypothesis when it's true, usually set at 0.05. It controls the risk of making a Type I error.

4.  What does a P-value represent in hypothesis testing?
 - The P-value represents the probability of observing the data, or something more extreme, if the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
 - If the P-value is less than the significance level, reject the null hypothesis; otherwise, fail to reject it.

6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type 1 error is rejecting a true null hypothesis, while Type 2 error is failing to reject a false null hypothesis.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 - A one-tailed test checks for an effect in one direction, while a two-tailed test checks in both directions.

8. What is the Z-test, and when is it used in hypothesis testing?
 - The Z-test is used to test population means when the population standard deviation is known and the sample size is large.

9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?
 - Z-score = (sample mean − population mean) / (standard deviation / √n); it shows how many standard deviations the sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
 - The T-distribution is a probability distribution used when the sample size is small and the population standard deviation is unknown.

11. What is the difference between a Z-test and a T-test?
 - Z-test is used with large samples and known population standard deviation; T-test is used with small samples and unknown population standard deviation.

12. What is the T-test, and how is it used in hypothesis testing?
 - The T-test compares sample means to assess if they differ significantly when population variance is unknown.

13. What is the relationship between Z-test and T-test in hypothesis testing?
 - Both test differences in means; Z-test applies with large samples and known variance, T-test applies with small samples and unknown variance.

14. What is a confidence interval, and how is it used to interpret statistical results?
 - A confidence interval estimates a range of values likely containing the population parameter, indicating result precision and reliability.

15.  What is the margin of error, and how does it affect the confidence interval?
 - Margin of error shows the range of possible sampling error and widens or narrows the confidence interval accordingly.

16.  How is Bayes' Theorem used in statistics, and what is its significance?
 - Bayes' Theorem updates the probability of an event based on new evidence, crucial for decision-making under uncertainty.

17. What is the Chi-square distribution, and when is it used?
 - The Chi-square distribution models the sum of squared standard normal variables and is used for testing independence and goodness-of-fit.

18. What is the Chi-square goodness of fit test, and how is it applied?
 - It tests if observed data fits an expected distribution by comparing frequencies using the Chi-square statistic.

19. What is the F-distribution, and when is it used in hypothesis testing?
 - The F-distribution compares variances and is used in ANOVA and tests of equality of variances.

20.  What is an ANOVA test, and what are its assumptions?
 - ANOVA tests for differences between means of three or more groups; it assumes normality, independence, and equal variances.

21. What are the different types of ANOVA tests?
 - Common types are one-way ANOVA, two-way ANOVA, and repeated measures ANOVA.

22. What is the F-test, and how does it relate to hypothesis testing?
 - The F-test compares variances to determine if group means are significantly different in hypothesis testing.

```python
3. #Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pmf(k, p):
    """
    Calculate the PMF of a Bernoulli distribution.
    k: outcome (0 or 1)
    p: probability of success (1)
    """
    if k not in [0, 1]:
        return 0
    return p**k * (1 - p)**(1 - k)

# Example usage
p_success = 0.7
print(f"P(X=1): {bernoulli_pmf(1, p_success)}")
print(f"P(X=0): {bernoulli_pmf(0, p_success)}")
```

```python
15. #Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([52, 55, 53, 58, 54, 56, 57, 59, 60, 54])
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Known population mean (null hypothesis)
pop_mean =  55

# Significance level
alpha = 0.05

# Calculate Z-statistic
z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Decision
if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
16. #Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

# Sample data
data = np.array([12, 15, 14, 16, 13, 14, 15, 17, 16, 14])
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
confidence = 0.95

# Calculate confidence interval
t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_error = t_crit * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print(f"Mean: {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
1. #Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
from math import sqrt
from scipy.stats import norm

# Sample data
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 30
significance_level = 0.05

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret result
if p_value < significance_level:
    print("Reject the null hypothesis: Sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
3. # Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
from math import sqrt
from scipy.stats import norm

def one_sample_z_test(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
    # Calculate Z-score
    z = (sample_mean - pop_mean) / (pop_std / sqrt(sample_size))
    # Calculate two-tailed p-value
    p = 2 * (1 - norm.cdf(abs(z)))
    # Print results
    print(f"Z-score: {z:.3f}")
    print(f"P-value: {p:.3f}")
    # Decision
    if p < alpha:
        print("Reject the null hypothesis: significant difference.")
    else:
        print("Fail to reject the null hypothesis: no significant difference.")

# Example usage
one_sample_z_test(sample_mean=102, pop_mean=100, pop_std=15, sample_size=40)
```

```python
4. #Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt

# Parameters
sample_mean = 104
pop_mean = 100
pop_std = 15
sample_size = 36
alpha = 0.05

# Calculate Z-score
z_score = (sample_mean - pop_mean) / (pop_std / sqrt(sample_size))

# Critical values for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# P-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Plot setup
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

# Mark calculated Z-score
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Labels
plt.title('Two-tailed Z-test Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()

plt.show()

# Print results
print(f"Z-score: {z_score:.3f}")
print(f"Critical Z-value: ±{z_critical:.3f}")
print(f"P-value: {p_value:.3f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
5. #Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from math import sqrt

def plot_type1_type2_errors(pop_mean=100, alt_mean=105, pop_std=15, sample_size=36, alpha=0.05):
    # Calculate critical z-value
    z_critical = norm.ppf(1 - alpha)
    # Calculate critical value in original units
    critical_value = pop_mean + z_critical * (pop_std / sqrt(sample_size))

    # Define x range for null and alternative distributions
    x = np.linspace(pop_mean - 4*pop_std, alt_mean + 4*pop_std, 1000)
    null_dist = norm.pdf(x, pop_mean, pop_std / sqrt(sample_size))
    alt_dist = norm.pdf(x, alt_mean, pop_std / sqrt(sample_size))

    plt.figure(figsize=(10,6))

    # Plot null hypothesis distribution
    plt.plot(x, null_dist, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative hypothesis distribution
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error region (α) - area under null distribution beyond critical value
    x_type1 = np.linspace(critical_value, x[-1], 100)
    plt.fill_between(x_type1, 0, norm.pdf(x_type1, pop_mean, pop_std / sqrt(sample_size)),
                     color='red', alpha=0.5, label='Type 1 Error (α)')

    # Shade Type 2 error region (β) - area under alternative distribution left of critical value
    x_type2 = np.linspace(x[0], critical_value, 100)
    plt.fill_between(x_type2, 0, norm.pdf(x_type2, alt_mean, pop_std / sqrt(sample_size)),
                     color='orange', alpha=0.5, label='Type 2 Error (β)')

    # Plot critical value line
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example call
plot_type1_type2_errors()
```

```python
9. #Write a Python function to calculate the confidence interval for a sample mean and explain its significance
from math import sqrt
from scipy.stats import t

def confidence_interval(sample_data, confidence=0.95):
    n = len(sample_data)
    mean = sum(sample_data) / n
    std_err = (sum((x - mean)**2 for x in sample_data) / (n - 1))**0.5 / sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_error = t_critical * std_err
    lower = mean - margin_error
    upper = mean + margin_error
    return (lower, upper)

# Example usage
data = [12, 15, 14, 16, 15, 13, 14]
ci = confidence_interval(data)
print(f"95% confidence interval: {ci}")

```

```python
10. #Write a Python program to calculate the margin of error for a given confidence level using sample data
from math import sqrt
from scipy.stats import t

def margin_of_error(sample_data, confidence=0.95):
    n = len(sample_data)
    mean = sum(sample_data) / n
    std_err = (sum((x - mean)**2 for x in sample_data) / (n - 1))**0.5 / sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    return t_critical * std_err

# Example usage
data = [10, 12, 15, 14, 13, 16, 12]
moe = margin_of_error(data)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

```python
11. #Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate posterior probability using Bayes' Theorem:
    posterior = (likelihood * prior) / evidence
    """
    return (likelihood * prior) / evidence

# Example:
# Suppose:
# Prior probability of disease: 1% (0.01)
# Likelihood of positive test if disease present (sensitivity): 99% (0.99)
# Probability of positive test overall (evidence): 5% (0.05)

prior = 0.01
likelihood = 0.99
evidence = 0.05

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior probability of disease given positive test: {posterior:.3f}")
```

```python
12. #Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table (rows: Gender, columns: Preference)
#       Like   Dislike
# Male   30      10
# Female 20      20

data = np.array([[30, 10],
                 [20, 20]])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-square statistic: {chi2:.3f}")
print(f"P-value: {p:.3f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject null hypothesis: variables are dependent.")
else:
    print("Fail to reject null hypothesis: variables are independent.")
```

```python
13. #Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data
import numpy as np

def calculate_expected_frequencies(observed):
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()
    expected = row_sums @ col_sums / total
    return expected

# Example observed data
observed = np.array([[30, 10],
                     [20, 20]])

expected = calculate_expected_frequencies(observed)
print("Expected frequencies:")
print(expected)
```
